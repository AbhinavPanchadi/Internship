```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# loading the dataset to a Pandas DataFrame
data = pd.read_csv('/content/creditcard.csv')
```

```python
# first 5 rows of the dataset
data.head()
```

```python
data.tail()
```

```python
# dataset informations
data.info()
```

```python
# checking the number of missing values in each column
data.isnull().sum()
```

```python
# distribution of legit transactions & fraudulent transactions
data['Class'].value_counts()
```

```python
# separating the data for analysis
legit = data[data.Class == 0]
fraud = data[data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the values for both transactions
data.groupby('Class').mean()
```

```python
legit_sample = legit.sample(n=492)
```

## Concatenating two DataFrames

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

## Splitting the data into Features & Targets

```python
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

## Split the data into Training data & Testing Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

## Model Training

## Logistic Regression

```python
model = LogisticRegression()
```

```python
# training the Logistic Regression Model with Training Data
model.fit(X_train, Y_train)
```

## Model Evaluation

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on Test Data : ', test_data_accuracy)
```

## Model Training 2

## SVM

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
# Initialize the SVM model
svm_model = SVC(kernel='linear')
```

```python
# Train the SVM model with training data
svm_model.fit(X_train, Y_train)
```

## Model Evaluation 2

```python
# Accuracy on training data
X_train_prediction_svm = svm_model.predict(X_train)
training_data_accuracy_svm = accuracy_score(X_train_prediction_svm, Y_train)
print('Accuracy on Training data (SVM):', training_data_accuracy_svm)
```

```python
# Accuracy on test data
X_test_prediction_svm = svm_model.predict(X_test)
test_data_accuracy_svm = accuracy_score(X_test_prediction_svm, Y_test)
print('Accuracy score on Test Data (SVM):', test_data_accuracy_svm)
```

# Comparision between two models

```python
import matplotlib.pyplot as plt
```

```python
# Labels for the models
labels = ['Logistic Regression - Train', 'Logistic Regression - Test',
          'SVM - Train', 'SVM - Test']
```

```python
# Accuracy values
accuracies = [training_data_accuracy, test_data_accuracy,
              training_data_accuracy_svm, test_data_accuracy_svm]
```

```python
# Colors for different sections
colors = ['blue', 'lightblue', 'green', 'lightgreen']
```

```python
# Create pie chart
plt.figure(figsize=(7, 7))
plt.pie(accuracies, labels=labels, autopct='%1.1f%%', colors=colors, startangle=90)

# Title
plt.title('Accuracy Distribution of Logistic Regression and SVM')

# Show the chart
plt.show()
```
