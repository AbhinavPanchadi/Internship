<center><h1>Handling missing data using Replace</center>

### single spatial values

```python
import numpy as np
import pandas as pd 
df=pd.read_csv(r"C:\Users\Abhinav\OneDrive\Desktop\jupyter projects\pandas replace\weather_data.csv")
df
```

```python
new_df=df.replace(-99999,np.nan)
new_df
```

### two spatial values

```python
df=pd.read_csv(r"C:\Users\Abhinav\OneDrive\Desktop\jupyter projects\pandas replace\weather_data_two_spatial_values.csv")
df
```

```python
new_df=df.replace([-99999,-8888],np.nan)
new_df
```

### replace column by specific value

```python
new_df=df.replace({
    'temperature':-99999,
    'windspeed':[-99999,-8888],
    'event':'0',
},np.nan)
new_df
```

### replace using mapping

```python
new_df = df.replace({
        -99999: np.nan,
        '0': 'Sunny',
    })
new_df
```

### regx (regular expression)

```python
df=pd.read_csv(r"C:\Users\Abhinav\OneDrive\Desktop\jupyter projects\pandas replace\weather_data_regx.csv")
df
```

```python
new_df=df.replace('[A-Za-z]','',regex=True) # it removed the non-numeric
new_df
                
```

```python
new_df = df.replace({'temperature': '[A-Za-z]', 'windspeed': '[a-z]'},'', regex=True) 
new_df
```

## replacing list with another list

```python
df = pd.DataFrame({
    'student': ['abhi', 'tillu', 'shek', 'revi', 'uday', 'sukku'],
    'score': ['exceptional','average', 'good', 'poor', 'average', 'exceptional']})
df
```

```python
df.replace(['poor', 'average', 'good', 'exceptional'], [1,2,3,4])
```
