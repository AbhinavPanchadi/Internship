```python
import pandas as pd
df=pd.read_csv(r"C:\Users\Abhinav\OneDrive\Desktop\jupyter projects\Time series analysis\DateRange\aapl_no_dates.csv")
df.head()
```

```python
range=pd.date_range(start="6/1/2017",end="6/30/2017",freq='B') # here B means Business days
range
```

```python
df.set_index(range,inplace=True) 
df
```

```python
%matplotlib inline
df.Close.plot()
```
