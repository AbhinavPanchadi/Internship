<center><h1>Concatenation</h1></center>

```python
import pandas as pd
india_weather=pd.DataFrame({
    "city":["mumbai","delhi","hyd"],
    "temp":[36,45,60],
    "humidity":[80,60,70]
})
india_weather
```

```python
us_weather=pd.DataFrame({
    "city":["vegas","nyc","LA"],
    "temp":[26,35,30],
    "humidity":[40,50,60]
})
us_weather
```

```python
df=pd.concat([india_weather,us_weather])
df
```

```python
df=pd.concat([india_weather,us_weather],ignore_index=True) #The index is being ignored and is in order 
df
```

```python
df=pd.concat([india_weather,us_weather],keys=["india","us"])
df
```

```python
df.loc['india']
```

```python
df.loc['us']
```

```python
temperature_df=pd.DataFrame({
    "city":["mumbai","delhi","hyd"],
    "temp":[36,45,60],
})
temperature_df
```

```python
windspeed_df=pd.DataFrame({
    "city":["mumbai","delhi","hyd"],
    "windspeed":[12,20,30],
})
windspeed_df
```

```python
df=pd.concat([temperature_df,windspeed_df])
df
```

```python
df=pd.concat([temperature_df,windspeed_df],axis=1)
df
```

#### missing data 

```python
windspeedd_df=pd.DataFrame({
    "city":["delhi","mumbai"],
    "windspeed":[12,20],
})
windspeedd_df
```

```python
df=pd.concat([temperature_df,windspeedd_df],axis=1)
df
```

```python
temperature_df=pd.DataFrame({
    "city":["mumbai","delhi","hyd"],
    "temp":[36,45,60],
},index=[0,1,2]) #assigning indexes helps us to put the data in ordered form

windspeedd_df=pd.DataFrame({
    "city":["delhi","mumbai"],
    "windspeed":[12,20],
},index=[1,0])
df=pd.concat([temperature_df,windspeedd_df],axis=1)
df
```

```python
s=pd.Series(["humid","dry","rainy"],name="event")
s
```

```python
df=pd.concat([temperature_df,s],axis=1)
df
```
