## loc and iloc

```python
np.random.seed(101)
array=np.random.randint(10,50,(10,5))
df=pd.DataFrame(data=array,index="A B C D E F G H I J".split(),columns="VAL1 VAL2 VAL3 VAL4 VAL5".split())
```

```python
df
```

```python
df.loc["A":"D"]
```

```python
df.iloc[0:4]
```

```python
int(df.loc["D","VAL4"])
```

```python
df.loc["C":"G",["VAL2"]]
```

```python
df.loc["C":"G"][["VAL2"]]
```

```python
df.loc["C":"G","VAL1":"VAL3"]
```

```python
df > 20
```

```python
df[df>20]
```

```python
df["VAL1"]<20
```

```python
df[df["VAL1"]<20]
```
